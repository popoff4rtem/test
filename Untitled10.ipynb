```python
import os
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.animation import ArtistAnimation




parmas = []

top = r'/Users/artempopov/Downloads/ellpced_scripts_1_20230707_222346'
for root, dirs, files in os.walk(top, topdown=False):
    for name in files:
        if name.find('monitor_dxy') > -1:
            parmas.append(name)


parmas = sorted(parmas)          #  
parmas.insert(0, parmas[-2])     #    (0- )
parmas.pop(-2)                   #   

fig  = plt.figure(figsize=(10,6))
ax_3d = fig.add_subplot(projection='3d')

frames = []

for i in parmas:
    data = '/Users/artempopov/Downloads/ellpced_scripts_1_20230707_222346/' + i 
    with open(f'{data}', 'r') as f:
      content = f.readlines()

    matrix = []
    for line in content[29:]:
      if line.find('# Errors') != -1:
        break
      else:
        raw = line.split()
        matrix.append([float(number) for number in raw])

    x = np.arange(-10,10,20/len(matrix))
    y = np.arange(-10,10,20/len(matrix))

    xgrid, ygrid = np.meshgrid(x,y)

    intensivity= np.array(matrix)
    
    line = ax_3d.plot_surface(xgrid, ygrid, intensivity, rstride = 3, cstride = 3, cmap="jet")
    frames.append([line])

    
animation = ArtistAnimation(
    fig,
    frames,
    interval = 200,
    blit = True,
    repeat = True)


plt.show()
```
